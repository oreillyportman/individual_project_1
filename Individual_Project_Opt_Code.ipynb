```python
from datetime import timedelta, date
import os

def Load_Data(file,single_legn,list_legn): #Loads project task list and/or project members data in.
    f=open(file, "r")
    contents = f.readlines()
    f.close()
    Members_Data = []
    Member_Data = [0]*single_legn
    for i in range(len(contents)):
        contents[i] = contents[i].strip('\n')
    for k in range(int((len(contents))/(single_legn + list_legn))):    
        for l in range(single_legn):
            Member_Data[l] = contents[l + k*(single_legn + list_legn)]
        for j in range(single_legn, (single_legn + list_legn)):     
            Member_Data.append([])
            set_a = 0
            for i in range(len(contents[j + k*(single_legn + list_legn)])):
                if contents[j + k*(single_legn + list_legn)][i] == ',':     
                    Member_Data[j].append(contents[j + k*(single_legn + list_legn)][set_a:i])
                    set_a = i+1  
        Members_Data.append(Member_Data)
        Member_Data = [0]*single_legn
    return Members_Data

def Dates_logic(Unavb_sort,i):
    for j in range(i+1,len(Unavb_sort)):
        if int(Unavb_sort[i][1][6:10]) > int(Unavb_sort[j][1][6:10]):
            Unavb_sort[i],Unavb_sort[j] = Unavb_sort[j],Unavb_sort[i]
        elif int(Unavb_sort[i][1][6:10]) == int(Unavb_sort[j][1][6:10]):
            if int(Unavb_sort[i][1][3:5]) > int(Unavb_sort[j][1][3:5]):
                Unavb_sort[i],Unavb_sort[j] = Unavb_sort[j],Unavb_sort[i]
            elif int(Unavb_sort[i][1][3:5]) == int(Unavb_sort[j][1][3:5]):
                if int(Unavb_sort[i][1][0:2]) > int(Unavb_sort[j][1][0:2]):
                    Unavb_sort[i],Unavb_sort[j] = Unavb_sort[j],Unavb_sort[i]

def Sort_Members(Mem_Data): #Sorts members by the start or end of their off data.
    Unavb = []
    Unavb_sort = []
    if len(Mem_Data[0]) == 6:
        for i in range(len(Mem_Data)):
            for j in range(len(Mem_Data[i][2])):
                Unavb.append([Mem_Data[i][0]])
                Unavb.append(Mem_Data[i][2][j])
                Unavb.append(Mem_Data[i][3][j])
        for i in range(int(len(Unavb)/3)):
            Unavb_sort.append([])   
            for j in range(0+i*3,3+i*3):
                Unavb_sort[i].append(Unavb[j])
        for i in range(len(Unavb_sort)):
            Dates_logic(Unavb_sort,i)
    else:
        Unavb_sort = Mem_Data
        for i in range(len(Unavb_sort)):
            Dates_logic(Unavb_sort,i)        
    return Unavb_sort

def Member_data(Mem_Data):
    Mem_list = []
    for i in range(len(Mem_Data)):
        Mem_list.append(Mem_Data[i][0])
    return Mem_list

def Mass_Load_Data_Projects(file_path): #Loads data from all projects in folder "Test_projects" located in the current directory
    
    Project_Data = []
    path, dirs, files = next(os.walk(file_path))
    for i in range(len(files)):
        Project_Data.append([])    
        Project_Data[i] = Load_Data(file_path + "/Test_Project_" + str(i) + ".txt",3,3)
    return Project_Data

def apend_name_loop(Uavb_Sorted,i,b_1,b_2,b_3,a): #Appends name/names from entry i + b_3 to entry i + b_2
    for j in range(len(Uavb_Sorted[i+b_1][0])-a):
            Uavb_Sorted[i+b_2][0].append(Uavb_Sorted[i+b_3][0][j])
            
def break_down_1(Uavb_Sorted,i,a_1,a_2,a_3,a_4):
    if int(Uavb_Sorted[i][2][6:10]) == int(Uavb_Sorted[i+1][2][6:10]) and int(Uavb_Sorted[i][2][3:5]) == int(Uavb_Sorted[i+1][2][3:5]) and int(Uavb_Sorted[i][2][0:2]) == int(Uavb_Sorted[i+1][2][0:2]):
        Uavb_Sorted[i][2] = Uavb_Sorted[i+1][1]
        apend_name_loop(Uavb_Sorted,i,0,1,0,0)
        Uavb_Sorted = Sort_Members(Uavb_Sorted)
    else:
        apend_name_loop(Uavb_Sorted,i,a_1,a_2,a_1,0)     
        Uavb_Sorted[i+a_1][1] = Uavb_Sorted[i+a_2][2]
        Uavb_Sorted = Sort_Members(Uavb_Sorted)
    return Uavb_Sorted

#def break_down_2_logic(a_1,a_2,a_3)
#    Uavb_Sorted[i][1+a_1]=Uavb_Sorted[i][2-a_1]    
#    Uavb_Sorted[i][2]=Uavb_Sorted[i][1]
#    apend_name_loop(Uavb_Sorted,i,0,2,0,0)
#    apend_name_loop(Uavb_Sorted,i,0,1,0,0)
#    Uavb_Sorted = Sort_Members(Uavb_Sorted)
    
def break_down_2(Uavb_Sorted,i,a):
    if a == 0:
        Uavb_Sorted.insert(i+2,[[],0,0])      
        Uavb_Sorted[i+2][1]=Uavb_Sorted[i+1][2]
        Uavb_Sorted[i+2][2]=Uavb_Sorted[i][2]
        
        Uavb_Sorted[i][2]=Uavb_Sorted[i+1][1]    
        Uavb_Sorted[i+1][2]=Uavb_Sorted[i+2][1]
        apend_name_loop(Uavb_Sorted,i,0,2,0,0)
        apend_name_loop(Uavb_Sorted,i,0,1,0,0)
        Uavb_Sorted = Sort_Members(Uavb_Sorted)
    else:
        Uavb_Sorted.insert(i+2,[[],0,0])
        Uavb_Sorted[i][2],Uavb_Sorted[i+1][1] = Uavb_Sorted[i+1][1],Uavb_Sorted[i][2]
        Uavb_Sorted[i+2][1],Uavb_Sorted[i+1][1] = Uavb_Sorted[i+1][1],Uavb_Sorted[i+2][1]
        Uavb_Sorted[i+2][2],Uavb_Sorted[i+1][2] = Uavb_Sorted[i+1][2],Uavb_Sorted[i+2][2]
        
        Uavb_Sorted[i+1][1] = Uavb_Sorted[i][2]
        Uavb_Sorted[i+1][2] = Uavb_Sorted[i+2][1]
        apend_name_loop(Uavb_Sorted,i,1,2,1,0)
        apend_name_loop(Uavb_Sorted,i,0,1,0,0)
        Uavb_Sorted = Sort_Members(Uavb_Sorted)
    return Uavb_Sorted

def Fuse_And_Split(z,Uavb_Sorted,start_date):
    R = 0
    while R < z:#2*len(Uavb_Sorted) - 1:
        R += 1
        for i in range(len(Uavb_Sorted)-1):
            if int(Uavb_Sorted[i][1][6:10]) == int(Uavb_Sorted[i+1][1][6:10]) and int(Uavb_Sorted[i][1][3:5]) == int(Uavb_Sorted[i+1][1][3:5]) and int(Uavb_Sorted[i][1][0:2]) == int(Uavb_Sorted[i+1][1][0:2]):               
                if int(Uavb_Sorted[i][2][6:10]) == int(Uavb_Sorted[i+1][2][6:10]) and int(Uavb_Sorted[i][2][3:5]) == int(Uavb_Sorted[i+1][2][3:5]) and int(Uavb_Sorted[i][2][0:2]) == int(Uavb_Sorted[i+1][2][0:2]):
                    apend_name_loop(Uavb_Sorted,i,1,0,1,0)
                    Uavb_Sorted.remove(Uavb_Sorted[i+1])
                    break
                elif int(Uavb_Sorted[i][2][6:10]) < int(Uavb_Sorted[i+1][2][6:10]):
                    Uavb_Sorted=break_down_1(Uavb_Sorted,i,1,0,0,1) 
                    break      
                elif int(Uavb_Sorted[i][2][6:10]) == int(Uavb_Sorted[i+1][2][6:10]):
                    if int(Uavb_Sorted[i][2][3:5]) < int(Uavb_Sorted[i+1][2][3:5]):
                        Uavb_Sorted=break_down_1(Uavb_Sorted,i,1,0,0,1)
                        break
                    elif int(Uavb_Sorted[i][2][3:5]) == int(Uavb_Sorted[i+1][2][3:5]): 
                        if int(Uavb_Sorted[i][2][0:2]) < int(Uavb_Sorted[i+1][2][0:2]):
                            Uavb_Sorted=break_down_1(Uavb_Sorted,i,1,0,0,1)
                            break 
                        else:
                            Uavb_Sorted=break_down_1(Uavb_Sorted,i,0,1,1,0)
                            break 
                    else:
                        Uavb_Sorted=break_down_1(Uavb_Sorted,i,0,1,1,0)
                        break
                else:
                    Uavb_Sorted=break_down_1(Uavb_Sorted,i,0,1,0,1)
                    break
            elif int(Uavb_Sorted[i][2][6:10]) == int(Uavb_Sorted[i+1][1][6:10]):
                if int(Uavb_Sorted[i][2][3:5]) == int(Uavb_Sorted[i+1][1][3:5]):
                    if int(Uavb_Sorted[i][2][0:2]) == int(Uavb_Sorted[i+1][1][0:2]):
                        if i == len(Uavb_Sorted)-1:
                            break
                    elif int(Uavb_Sorted[i][2][0:2]) > int(Uavb_Sorted[i+1][1][0:2]) and int(Uavb_Sorted[i][2][0:2]) > int(Uavb_Sorted[i+1][2][0:2]):
                        if int(Uavb_Sorted[i][2][6:10]) == int(Uavb_Sorted[i+1][2][6:10]):
                            Uavb_Sorted=break_down_2(Uavb_Sorted,i,0)
                            break
                        else:
                            Uavb_Sorted=break_down_2(Uavb_Sorted,i,1)
                            break
                    elif int(Uavb_Sorted[i][2][0:2]) > int(Uavb_Sorted[i+1][1][0:2]) and int(Uavb_Sorted[i][2][0:2]) < int(Uavb_Sorted[i+1][2][0:2]):
                        Uavb_Sorted=break_down_2(Uavb_Sorted,i,1)
                        break
                    elif int(Uavb_Sorted[i][2][6:10]) == int(Uavb_Sorted[i+1][2][6:10]) and int(Uavb_Sorted[i][2][3:5]) == int(Uavb_Sorted[i+1][2][3:5]) and int(Uavb_Sorted[i][2][0:2]) == int(Uavb_Sorted[i+1][2][0:2]):
                        Uavb_Sorted[i][2] = Uavb_Sorted[i+1][1]
                        apend_name_loop(Uavb_Sorted,i,0,1,0,0)
                        Uavb_Sorted = Sort_Members(Uavb_Sorted)
                        break
                elif int(Uavb_Sorted[i][2][3:5]) > int(Uavb_Sorted[i+1][1][3:5]):                    
                    if int(Uavb_Sorted[i][2][6:10]) == int(Uavb_Sorted[i+1][2][6:10]) and int(Uavb_Sorted[i][2][3:5]) == int(Uavb_Sorted[i+1][2][3:5]) and int(Uavb_Sorted[i][2][0:2]) == int(Uavb_Sorted[i+1][2][0:2]):
                        Uavb_Sorted[i][2] = Uavb_Sorted[i+1][1]
                        apend_name_loop(Uavb_Sorted,i,0,1,0,0)
                        Uavb_Sorted = Sort_Members(Uavb_Sorted)
                        break           
                    elif int(Uavb_Sorted[i][2][0:2]) < int(Uavb_Sorted[i+1][2][0:2]):
                        Uavb_Sorted=break_down_2(Uavb_Sorted,i,1)
                    else:
                        Uavb_Sorted=break_down_2(Uavb_Sorted,i,0)
                else:
                    Uavb_Sorted=break_down_2(Uavb_Sorted,i,0)
                    break
            elif int(Uavb_Sorted[i][2][6:10]) > int(Uavb_Sorted[i+1][1][6:10]):
                if int(Uavb_Sorted[i][2][6:10]) == int(Uavb_Sorted[i+1][2][6:10]) and int(Uavb_Sorted[i][2][3:5]) == int(Uavb_Sorted[i+1][2][3:5]) and int(Uavb_Sorted[i][2][0:2]) == int(Uavb_Sorted[i+1][2][0:2]):
                    Uavb_Sorted[i][2] = Uavb_Sorted[i+1][1]
                    apend_name_loop(Uavb_Sorted,i,0,1,0,0)
                    Uavb_Sorted = Sort_Members(Uavb_Sorted)
                    break 
                elif int(Uavb_Sorted[i][2][6:10]) == int(Uavb_Sorted[i+1][2][6:10]) and int(Uavb_Sorted[i][2][3:5]) == int(Uavb_Sorted[i+1][2][3:5]) and int(Uavb_Sorted[i][2][0:2]) < int(Uavb_Sorted[i+1][2][0:2]):
                    Uavb_Sorted=break_down_2(Uavb_Sorted,i,1)
                    break
                elif int(Uavb_Sorted[i][2][6:10]) == int(Uavb_Sorted[i+1][2][6:10]) and int(Uavb_Sorted[i][2][3:5]) == int(Uavb_Sorted[i+1][2][3:5]) and int(Uavb_Sorted[i][2][0:2]) > int(Uavb_Sorted[i+1][2][0:2]):    
                    Uavb_Sorted=break_down_2(Uavb_Sorted,i,0)
                    break
                elif int(Uavb_Sorted[i][2][6:10]) == int(Uavb_Sorted[i+1][2][6:10]) and int(Uavb_Sorted[i][2][3:5]) < int(Uavb_Sorted[i+1][2][3:5]):
                    Uavb_Sorted=break_down_2(Uavb_Sorted,i,1)
                    break
                elif int(Uavb_Sorted[i][2][6:10]) == int(Uavb_Sorted[i+1][2][6:10]) and int(Uavb_Sorted[i][2][3:5]) > int(Uavb_Sorted[i+1][2][3:5]):                    
                    Uavb_Sorted=break_down_2(Uavb_Sorted,i,0)
                    break
                else:
                    if int(Uavb_Sorted[i][2][6:10]) > int(Uavb_Sorted[i+1][2][6:10]):
                        Uavb_Sorted=break_down_2(Uavb_Sorted,i,0)
                        break
                    else:
                        Uavb_Sorted=break_down_2(Uavb_Sorted,i,1)
                        break
    Uavb_Sorted.insert(0,[[],start_date,Uavb_Sorted[0][1]])#Give option for start date
    return Uavb_Sorted

def Continuum_logic(Uavb_Sorted,a,b):
    if Uavb_Sorted[i][2][a:b] != Uavb_Sorted[i+1][1][a:b]:
            Uavb_Sorted.insert(i+1,[[],Uavb_Sorted[i][2],Uavb_Sorted[i+1][1]])
    return Uavb_Sorted
            
def Continumm_quy(Uavb_Sorted):
    for i in range(3):
        Continuum_logic(Uavb_Sorted,i*3,(i+1)^2 + 1)
    return Uavb_Sorted

        
def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)

def date_gen(start_date, end_date):
    dates = []
    for single_date in daterange(start_date, end_date):
        dates.append(single_date.strftime("%d/%m/%Y")) 
    return dates

def Data_Scrub(Project_Data,Init_Start):
    fixed = []
    for i in range(len(Project_Data)):
        for j in range(len(Project_Data[i])):
            Project_Data[i][j][1] = int(Project_Data[i][j][1])
            if Project_Data[i][j][5][0] == 'N/A':         
                del Project_Data[i][j][5][0]
                Project_Data[i][j].insert(2,Init_Start)
                Project_Data[i][j].append('0')
                Project_Data[i][j].append(str(i + 1))
            else:
                Project_Data[i][j].insert(2,'0')
                Project_Data[i][j].append('0')
                Project_Data[i][j].append(str(i + 1))
    return Project_Data

def Opt_step_A(Uavb_Sorted, Mem_Data, Project_Data,i,j,k):
    b = 0
    c = 0
    Opt_Log = []
    Skill_Level = ['Novice', 'Average', 'Expericenced', 'Expert']
    for l in range(len(Uavb_Sorted[k])):  
        for m in range(len(Uavb_Sorted[k][0])):
            for n in range(len(Mem_Data)):
                b = 0                            
                if Uavb_Sorted[k][0][m] == Mem_Data[n][0]:
                    if set(Project_Data[i][j][4]).issubset(set(Mem_Data[n][4])) == True:                     
                        for p in range(len(Project_Data[i][j][4])):
                            if Project_Data[i][j][5][p] == Mem_Data[n][5][Mem_Data[n][4].index(Project_Data[i][j][4][p])]:      
                                b += 1
                            elif Skill_Level.index(Project_Data[i][j][5][p]) < Skill_Level.index(Mem_Data[n][5][Mem_Data[n][4].index(Project_Data[i][j][4][p])]):                                            
                                b += 1
                        if b == len(Project_Data[i][j][4]):    
                            Opt_Log.append([])
                            Opt_Log[c].append(Mem_Data[n][0])
                            Opt_Log[c].append(Mem_Data[n][1][0])
                            c += 1              
        if l != len(Uavb_Sorted[k])-1:
            Opt_Log = []
            c = 0
    return [Opt_Log,Uavb_Sorted[k][1:4],Project_Data[i][j][1:4] + Project_Data[i][j][7:9]]

def Storing(Store, Uavb_Sorted, Mem_Data, Project_Data,i,k):
    for j in range(len(Project_Data[i])):      
        if len(Project_Data[i][j][6]) == 0 and Project_Data[i][j][7] == '0':         
            Store.append(Opt_step_A(Uavb_Sorted, Mem_Data, Project_Data,i,j,k) + [j])
            
def Opt_Step_B_Single_Logic(Project_Data,Store,i,l,format_time_new,L):
    Project_Data[i][int(Store[l][3])][1] = float(Project_Data[i][int(Store[l][3])][1]) - float(Store[l][0][L][1])
    Project_Data[i][int(Store[l][3])][7] = "1"
    Project_Data[i][int(Store[l][3])][3] = format_time_new[len(format_time_new) - 1]
    for z in range(len(Project_Data[i])):
        if len(Project_Data[i][z][6]) > 0:
            for y in range(len(Project_Data[i][z][6])):
                if Project_Data[i][z][6][y] == Project_Data[i][int(Store[l][3])][0]:
                    del Project_Data[i][z][6][y]
                    break
            if len(Project_Data[i][z][6]) == 0:
                Project_Data[i][z][2] = Project_Data[i][int(Store[l][3])][3]
```

```python
def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)

def date_gen(start_date, end_date):
    dates = []
    for single_date in daterange(start_date, end_date):
        dates.append(single_date.strftime("%d/%m/%Y")) 
    return dates
```
